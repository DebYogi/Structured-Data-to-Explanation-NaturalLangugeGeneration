```python
# Manual Run: Tiny E2E Experiment
#
# This notebook demonstrates running the tiny end-to-end experiment and inspecting outputs.

# 1) Run the tiny experiment (creates outputs under outputs/example_run_notebook/)
from seq2seq_sar.experiments import run_tiny_end_to_end
out_dir = run_tiny_end_to_end(n_accounts=10, epochs=0, model="t5-small", seed=1, out_dir="outputs/example_run_notebook")
print(f"Experiment outputs -> {out_dir}")

# 2) Inspect produced files
from pathlib import Path
import json
p = Path(out_dir)
print("Summary exists:", (p / "summary.json").exists())
print("Pipeline files:", list((p / "pipeline").glob("sar_*.json")))

# Display one SAR JSON
sample = list((p / "pipeline").glob("sar_*.json"))[0]
print(sample)
print(json.loads(sample.read_text()))
```
