```python
# Update Workflows helper

# This notebook demonstrates how to programmatically write GitHub workflow YAML files and validate them.

import yaml
from pathlib import Path

example = {
    'name': 'example',
    'on': {'workflow_dispatch': {}},
    'jobs': {'noop': {'runs-on': 'ubuntu-latest', 'steps': [{'run': 'echo hi'}]}}
}

out = Path('generated_workflow.yml')
out.write_text(yaml.safe_dump(example))
print('Wrote', out)

# Validate YAML syntax using yamllint or by re-loading
try:
    loaded = yaml.safe_load(out.read_text())
    print('YAML parsed OK')
except Exception as e:
    print('YAML parsing error', e)
```
